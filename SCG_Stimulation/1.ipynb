```python
import laserfun as lf
# Build Pulse
pulse1 = lf.Pulse(pulse_type='sech', center_wavelength_nm=1551, 
                  fwhm_ps=1.3*1.76, time_window_ps=100, power=1, power_is_avg=True, frep_MHz =507.614)

# Build Fibre
SMF_28 = lf.Fiber(length=10, center_wl_nm=1310, dispersion_format='GVD',
                 dispersion=[0.076e-3,0,0], gamma_W_m=2.3e-3, loss_dB_per_m=0)

DSF_1 = lf.Fiber(length=13, center_wl_nm=1492, dispersion_format='GVD',
                 dispersion=[0.03889e-3,0,0], gamma_W_m=2.3e-3, loss_dB_per_m=0)

DSF_2 = lf.Fiber(length=4.5, center_wl_nm=1546, dispersion_format='GVD',
                 dispersion=[0.05557e-3,0,0], gamma_W_m=2.3e-3, loss_dB_per_m=0)

HNL_DSF = lf.Fiber(length=4, center_wl_nm=1550, dispersion_format='GVD',
                 dispersion=[0.0265e-3,0,0], gamma_W_m=9.9e-3 , loss_dB_per_m=0)
# Get Result
result1 = lf.NLSE(pulse1, SMF_28 , print_status=False, raman=True)
result2 = lf.NLSE(result1.pulse_out, DSF_1 , print_status=False, raman=True)
result3 = lf.NLSE(result2.pulse_out, DSF_2 , print_status=False, raman=True)

# Plot Result
result1.plot(wavelength = True, margin =1)
result2.plot(wavelength = True, margin =1)
result3.plot(wavelength = True, margin =1)
```
