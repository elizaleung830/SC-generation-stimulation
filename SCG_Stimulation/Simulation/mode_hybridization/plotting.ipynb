```python
import matplotlib.pyplot as plt
import numpy as np

data = np.load("data_width_sweep_1550nm.npz")
neffs = [data['neff_list_te'],data['neff_list_te1'],data['neff_list_tm0']]
width_list = data['width_list']

fig, axs = plt.subplots(1, 1, figsize=(10,10))
axs.set_xlabel("top width (um)")
axs.set_ylabel("neff")
axs.set_title("top width of waveguide vs neff at wavelength = 1550nm")
axs.scatter(width_list,neffs[0], color= "orange", label = "te0 mode")
axs.scatter(width_list,neffs[1], color="red", label = "te1 mode")
axs.scatter(width_list,neffs[2], color="blue", label = "tm0 mode")
axs.legend()
```
