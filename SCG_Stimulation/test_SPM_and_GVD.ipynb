 Testing the combined effect of GVD and SPM under the condition:
 - In the regime where $L_d \approx L_{NL}$ or in another word $N \approx 1$
 - no raman scattering
 - no fiber loss. $L_{eff} = L$ 
 - Gaussian pulse
Some equations that will be used:
- $L_d = \frac{T_0^2}{\beta_2}$ 
- $L_{NL} = \frac{1}{\gamma P_0}$

```python
import laserfun as lf
# Build Pulse
pulse1 = lf.Pulse(pulse_type='gaussian', center_wavelength_nm=1550, 
                 fwhm_ps=0.2, time_window_ps=100, power=5)

# Build Fibre
fiber1 = lf.Fiber(length=1, center_wl_nm=1550, dispersion_format='GVD',
                 dispersion=[0.04,0,0], gamma_W_m=0.2, loss_dB_per_m=0)

# Get Result
results1 = lf.NLSE(pulse1, fiber1 , print_status=False, raman=False)
# Plot Result
results1.plot(margin =2)
```
