 Testing effect solely from self phase modulation under condition where there is
 - no GVD effect $\beta _ {2,3,4} = 0$
 - no raman scattering
 - no fiber loss. $L_{eff} = L$

Condition: 
- Gaussian pulse shape (m =1)
- $\gamma = 0.2 W^{-1}m^{-1}$
- $P_0 = 5W$
- $T_0 = 0.2ps$
- $L = 1m $

Prediction:
\begin{align}
L_{NL} &= \frac{1}{\gamma P_0} = 1m \\
L_{eff} &= L = 1m \\
\phi_{max} &= \gamma P_0 L_{eff} = 1 \\
\delta\omega &= 0.86\Delta\omega \phi_{max} = \frac{0.86\phi_{max}}{T_0} = 4.3THz
\end{align}
- little broadening of around 4.3 Thz 
- From $\phi_{max} \approx (M -\frac{1}{2})\pi$ only 1 peak formed in the spectrum 

```python
import laserfun as lf
# Build Pulse
pulse1 = lf.Pulse(pulse_type='gaussian', center_wavelength_nm=1550, 
                 fwhm_ps=0.2, time_window_ps=10, power=5)

# Build Fibre
fiber1 = lf.Fiber(length=1, center_wl_nm=1550, dispersion_format='GVD',
                 dispersion=[0,0,0], gamma_W_m=0.2, loss_dB_per_m=0)

# Get Result
results1 = lf.NLSE(pulse1, fiber1 , print_status=False, raman=False)
# Plot Result
results1.plot()
```

Now test different shape of pulse
Condition: 
- 1: Gaussian pulse shape (m =1) 2: sech shape 3: sinc shape
- $\gamma = 0.2 W^{-1}m^{-1}$
- $P_0 = 5W$
- $T_0 = 0.2ps$
- $L = 1m $

Prediction: 
- sharper pulse shape will have more broadening effect from the equation $\delta \omega = \frac{mf(m)}{T_0} \phi_{max}$ 

```python
import laserfun as lf
# Build Pulse
pulse1 = lf.Pulse(pulse_type='gaussian', center_wavelength_nm=1550, 
                 fwhm_ps=0.2, time_window_ps=10, power=5)
pulse2 = lf.Pulse(pulse_type='sech', center_wavelength_nm=1550, 
                 fwhm_ps=0.2, time_window_ps=10, power=5)
pulse3 = lf.Pulse(pulse_type='sinc', center_wavelength_nm=1550, 
                 fwhm_ps=0.2, time_window_ps=10, power=5)
# Build Fibre
fiber1 = lf.Fiber(length=1, center_wl_nm=1550, dispersion_format='GVD',
                 dispersion=[0,0,0], gamma_W_m=0.2, loss_dB_per_m=0)

# Get Result
results1 = lf.NLSE(pulse1, fiber1 , print_status=False, raman=False)
results2 = lf.NLSE(pulse2, fiber1 , print_status=False, raman=False)
results3 = lf.NLSE(pulse3, fiber1 , print_status=False, raman=False)
# Plot Result
results1.plot()
results2.plot()
results3.plot()
```

Qualitative observation (for gaussian pulse): 
- Spectrum broaden more with shorter pulse (number of peak un effected)
- Spectrum broaden more with increase length and power, but number of peak in spectrum increase as well
