# Understanding Propagation in a Waveguide

### In this code we define a waveguide and a gaussian pulse and see how the pulse travels through the waveguide

```python
# importing built-in libraries

import numpy as np
import matplotlib.pyplot as mplot
from scipy import signal
from scipy.fftpack import fft, ifft
from scipy.fftpack import fftshift, ifftshift
from scipy.interpolate import UnivariateSpline

# importing built-in library for diffraction and interference
from diffractio import um, nm, mm, np,degrees
from diffractio.scalar_fields_X import Scalar_field_X
from diffractio.scalar_sources_X import Scalar_source_X
from diffractio.scalar_masks_X import Scalar_mask_X
from diffractio.scalar_fields_XZ import Scalar_field_XZ
from diffractio.scalar_masks_XZ import Scalar_mask_XZ
```

### The refractive index of the waveguide

```python
x = np.linspace(-2 * um, 2 * um, 512)
wavelength = 1.55 * um

t0=Scalar_mask_X(x, wavelength)
t0.slit(x0=0, size=0.5*um)
t0.draw()
```

```python
x0 = np.linspace(-2 * um, 2 * um, 512)
z0 = np.linspace(0 * um, 50 * um, 512)
wavelength = 1.55 * um
u0 = Scalar_source_X(x=x0, wavelength=wavelength)
u0.gauss_beam(A=1, x0=0 * um, z0=0 * um, w0=0.50 * um, theta=0 * degrees)
u0.draw()

u1 = Scalar_mask_XZ(x=x0, z=z0, wavelength=wavelength)
u1.incident_field(u0)
u1.rectangle(r0=(0 * um, 25 * um),
  size=(0.5 * um, 40 * um),
  refraction_index=1)

u1.draw_refraction_index()
u1.BPM(verbose=True, division=False)
u1.draw(logarithm=True, normalize='maximum', draw_borders=True);
```

### For a smaller waveguide (chip-sized)

```python
x0 = np.linspace(-20* um, 20 * um, 512)
z0 = np.linspace(0 * um, 50 * um, 512)
wavelength = 1.55 * um
u0 = Scalar_source_X(x=x0, wavelength=wavelength)
u0.gauss_beam(A=1, x0=0 * um, z0=0 * um, w0=0.5 * um, theta=0 * degrees)
u0.draw()

u1 = Scalar_mask_XZ(x=x0, z=z0, wavelength=wavelength)
u1.incident_field(u0)
u1.rectangle(r0=(0 * um, 22 * um),
  size=(2 * um, 40 * um),
  refraction_index=2)

u1.draw_refraction_index()
u1.BPM(verbose=True, division=False)
u1.draw(logarithm=True, normalize='maximum', draw_borders=True);
```
